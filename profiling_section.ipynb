```python
# profiling section
processed = pd.read_csv('output.csv')
df = pd.read_excel('module_2.xlsx')
correct = df[df['correctness']==1]
incorrect = df[df['correctness']==0]
aggs = correct.groupby('shot').agg(['min', 'max', 'mean'])

def calculate(col_name, col_val, shot):
  global aggs
  if col_val <= aggs[col_name]['min'][shot] and col_val >= aggs[col_name]['max'][shot]:
    return -1
  else:
    return 1

for col in cols:
  processed[col] = df.apply(lambda row : calculate(col, row[col],row['shot']), axis = 1)


videos = {
    'back foot deffence' : 'https://www.youtube.com/watch?v=er1VOa39BmI&list=PL5jPUzJ7Q8wJtMPlHt9-OohA32UlBW9I1&index=4',
    'back foot drive' : ' https://www.youtube.com/watch?v=gV13F3v-K0s',
    'cut' : "https://www.youtube.com/watch?v=PVGYVtsPPgE",
    'front foot defence' : 'https://www.youtube.com/watch?v=pmlr0IcdSPk&list=PL5jPUzJ7Q8wJtMPlHt9-OohA32UlBW9I1&index=3',
    'front foot drive' : 'https://www.youtube.com/watch?v=gQNB7jT3GDQ',
    'pull': 'https://www.youtube.com/watch?v=Wv1aSL_ODlo',
    'sweep' : 'https://www.youtube.com/watch?v=-9wgQGSnSnI'
}

threshold = 5

def suggest_video(max_val, shot):
  global videos
  if int(max_val) > threshold:
    return videos[shot.strip()]
  else:
    return "N/A"

def print_row(row):
  print(row.name[1])
  return "ASD"

player_aggs = processed.groupby(['player','shot']).agg(['sum'], axis=1)
player_aggs['MAX'] = player_aggs[cols].max(axis=1)
player_aggs['Video'] = player_aggs.apply(lambda row : suggest_video(row['MAX'],row.name[1]), axis=1)
player_aggs.head(20)
```
