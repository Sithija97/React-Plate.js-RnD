```python
import pandas as pd
import numpy as np

cols=['fsv.st.ba','fsv.st.tha', 'fsv.st.bha', 'fsv.st.ff.ka', 'fsv.st.bf.ka',
       'fsv.st.hip.a', 'fsv.bl.ba', 'fsv.bl.tha', 'fsv.bl.bha', 'fsv.bl.ff.ka',
       'fsv.bl.bf.ka', 'fsv.bl.hip.a', 'fsv.ds.ba', 'fsv.ds.tha', 'fsv.ds.bha',
       'fsv.ds.ff.ka', 'fsv.ds.bf.ka', 'fsv.ds.hip.a', 'fsv.im.ba',
       'fsv.im.tha', 'fsv.im.bha', 'fsv.im.ff.ka', 'fsv.im.bf.ka',
       'fsv.im.hip.a', 'fsv.ft.ba', 'fsv.ft.tha', 'fsv.ft.bha', 'fsv.ft.ff.ka',
       'fsv.ft.bf.ka', 'fsv.ft.hip.a', 'bsv.ds.bha', 'bsv.ds.ff.ka',
       'bsv.ds.bf.ka', 'bsv.ds.hip.a', 'bsv.im.tha', 'bsv.im.bha',
       'bsv.im.ff.ka', 'bsv.im.bf.ka', 'bsv.im.hip.a', 'bsv.ft.ba',
       'bsv.ft.tha', 'bsv.ft.bha', 'bsv.ft.ff.ka', 'bsv.ft.bf.ka',
       'bsv.ft.hip.a']
       
df = pd.read_excel('module_2.xlsx')
correct = df[df['correctness']==1]
incorrect = df[df['correctness']==0]
aggs = correct.groupby('shot').agg(['min', 'max', 'mean'])
```

```python
def calculate(col_name, col_val, shot):
  global aggs
  if col_val >= aggs[col_name]['min'][shot] and col_val <= aggs[col_name]['max'][shot]:
    return 0
  else:
    return (aggs[col_name]['mean'][shot] - col_val)

for col in cols:
  incorrect[col] = df.apply(lambda row : calculate(col, row[col],row['shot']), axis = 1)

incorrect.to_csv('output.csv')
incorrect
```
